# Data Cleaning

## Introduction
The data of this data set is unsatisfactory for use as a customer call list. There are errors in the information as well as unnecessary columns and rows that would slow down any person who was using this list. Our goal here is to clean the data set so it is simple and quick to read and utilize.

### Import and View Data
The first way to know what to clean in a dataset is to figure out what you want to do with it. For this dataset we are preparing it to be used as a customer call list. Using this we can deduce that we should pay attention to the Do_Not_Contact column so we can drop rows where the customer is not being contacted, and the Not_Useful_Column because it holds no information that is relevant to the situaion.

The next way to know what we need to do with the data set is by looking through and finding inconsistant data. In this data set we see that Last_Name is not standardized, having multiple different characters that are not part of the name. We can also see that Phone_Number holds different formats for phone numbers. Paying Customer and Do_Not_Contact have the same issue where a Y/N answer holds multiple different ways of answering. Finally we can see that there are also duplicate rows.

We now know that these are the ways that we can clean and standardize the data for use, so lets begin!

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\ciret\OneDrive\Desktop\Data Analyst\Data\Customer Call List.xlsx")
df
```

### Duplicate Deletion
Normally we first search to see if there is duplicates in a dataset before executing this code, we say previously that there was already a duplicate row so there is no need to search. This is a simple way of removing duplicates from a dataset.

```python
df = df.drop_duplicates()
df
```

### Drop Not_Useful_Column
One of the columns that we have are not necessary for our call list so we can drop it with the following code.

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

### Standardize Last_Name
Next we can standardize last names so that all names hold the same format, making it easier to read, wrangle, and ensure it mates the customer's last name.

```python
df["Last_Name"] = df["Last_Name"].str.strip("123456789./_")
df
```

### Standardize Phone_Number
The next two lines are to standardize the phonenumbers and remove non-phone numbers from the column respectively.

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex = True)
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

### Expanding Address
Because Address is already formated we don't need to standardize it but we can use this to expand it into mupltple columns for later use if desired.

```python
df[["Street Adress", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2,expand=True)
df
```

### Standardize Paying Customer and Do_Not_Contact
With these to columns having the same error of nonstandardized data we can use similar code to unify them, ensuring that each row is some variation of Y or N

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df = df.fillna('')
df = df.replace('N/a','')
df
```

### Droping Unnecessary Rows
As a call list, we need a phone number to complete this action so we can drop rows that don't have a phone number. We can also drop rows that have a 'Y' under Do_Not_Contact because it means we are not meant to contact that customer.

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

### Final Touches
We have now standardized the data to where it's useable as a customer call list so we can index it properly. Executing the following code allows for a ordered index on the customer call list.

```python
df = df.reset_index(drop=True)
df
```
